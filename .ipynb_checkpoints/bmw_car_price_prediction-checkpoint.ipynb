```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
import seaborn as sns
import numpy as np
from scipy import stats


data = pd.read_csv('true_car_listings.csv')
data["NumOfYears"] = 2018 - data["Year"]
bmw_models = data[data['Make'] == 'BMW']['Model'].value_counts()[:34].index

bmw_data = data[(data['Make'] == 'BMW') & (data['Model'].isin(bmw_models))]
bmw_data
```

```python
# ÐŸÑ€Ð¾Ð²ÐµÑ€ÐºÐ° Ð½Ð° Ð½Ð¾Ñ€Ð¼Ð°Ð»ÑŒÐ½Ð¾ÑÑ‚ÑŒ

for col in ['Price', 'Mileage', 'NumOfYears']:
    var = bmw_data[col]


    print('\n' + col)
    d_stat, p_value_ks = stats.kstest(var, 'norm', args=(var.mean(), var.std()))
    print(p_value_ks)
    if p_value_ks > 0.05:
        print('Normal')
    else:
        print('Not Normal')

#â€¢ Ð˜Ð½Ñ‚ÐµÑ€Ð¿Ñ€ÐµÑ‚Ð°Ñ†Ð¸Ñ: P-value Ð²Ñ‹ÑˆÐµ 0.05, Ñ‡Ñ‚Ð¾ ÑƒÐºÐ°Ð·Ñ‹Ð²Ð°ÐµÑ‚ Ð½Ð° Ñ‚Ð¾, Ñ‡Ñ‚Ð¾
#Ñ€Ð°ÑÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ðµ Ð¾Ñ†ÐµÐ½Ð¾Ðº ÑÑ‚ÑƒÐ´ÐµÐ½Ñ‚Ð¾Ð² Ð½Ðµ Ð¾Ñ‚Ð»Ð¸Ñ‡Ð°ÐµÑ‚ÑÑ Ð¾Ñ‚ Ð½Ð¾Ñ€Ð¼Ð°Ð»ÑŒÐ½Ð¾Ð³Ð¾ Ñ€Ð°ÑÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ñ.
```

```python
# ÐŸÑ€Ð¾Ð²ÐµÑ€ÐºÐ° Ð½Ð° Ð¼ÑƒÐ»ÑŒÑ‚Ð¸ÐºÐ¾Ð»Ð»ÐµÐ½Ð¸Ð°Ñ€Ð½Ð¾ÑÑ‚ÑŒ

sns.pairplot(bmw_data)
```

```python
print(np.corrcoef(bmw_data['Mileage'], bmw_data['NumOfYears']) )

bmw_data[['Mileage', 'NumOfYears']].corr()

```

```python
import statsmodels.api as sm

X = bmw_data[['Mileage']]
X = sm.add_constant(X)
y = bmw_data['Price']

model = sm.regression.linear_model.OLS(y, X).fit()
print(model.summary())

# Residuals plot for homoscedasticity check
residuals = model.resid
plt.figure(figsize=(10, 6))
plt.scatter(model.predict(), residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()

plt.title('Residuals Distribution')
plt.hist(residuals)
plt.show()
```

```python
bmw_data['State'] = bmw_data['State'].apply(lambda x: x.upper().strip())

bmw_data[bmw_data['State'] == 'DC']
```

```python
import statsmodels.api as sm

pd.get_dummies(bmw_data['Model'])



X = bmw_data[['Mileage', 'NumOfYears']]
X = sm.add_constant(X)
X = X.join(pd.get_dummies(bmw_data['Model']))
X = X.join(pd.get_dummies(bmw_data['State']))
y = bmw_data['Price']

model = sm.regression.linear_model.OLS(y, X).fit()
print(model.summary())

# Residuals plot for homoscedasticity check
residuals = model.resid
plt.figure(figsize=(10, 6))
plt.scatter(model.predict(), residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()

plt.title('Residuals Distribution')
plt.hist(residuals)
plt.show()
```

```python
good_cols = [col for col, pvalue in list(zip(model.pvalues.index, model.pvalues.values)) if pvalue < 0.05]
```

```python
import statsmodels.api as sm

pd.get_dummies(bmw_data['Model'])



X = bmw_data[['Mileage', 'NumOfYears']]
X = sm.add_constant(X)
X = X.join(pd.get_dummies(bmw_data['Model']))
X = X.join(pd.get_dummies(bmw_data['State']))
X = X[good_cols]
y = bmw_data['Price']

model = sm.regression.linear_model.OLS(y, X).fit()
print(model.summary())

# Residuals plot for homoscedasticity check
residuals = model.resid
plt.figure(figsize=(10, 6))
plt.scatter(model.predict(), residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()

plt.title('Residuals Distribution')
plt.hist(residuals)
plt.show()
```

```python
from statsmodels.formula.api import ols
from scipy import stats

# Get model residuals
residuals = model.resid

# Scatter plot for residuals (Visual check for homoscedasticity)
plt.scatter(model.predict(), residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Values')
plt.show()

# Breusch-Pagan test for homoscedasticity
_, bp_pvalue, _, f_pvalue = sm.stats.diagnostic.het_breuschpagan(residuals, model.model.exog)
print('Breusch-Pagan test p-value:', bp_pvalue)

# Interpretation of p-value
# Typically, a p-value less than 0.05 suggests heteroscedasticity
```

## Train, test, validation split

```python
from sklearn.model_selection import train_test_split

# Define the size of the train, test, and validation sets
train_size = 0.7  # 70% of the data for training
test_size = 0.15  # 15% of the data for testing
val_size = 0.15   # 15% of the data for validation

# Split the list of dataframes into train and test sets
train_data, test_data = train_test_split(df, test_size=test_size, random_state=42)

# Split the test set further into test and validation sets
test_data, val_data = train_test_split(test_data, test_size=val_size/(test_size+val_size), random_state=42)
```

```python
train_data.shape, test_data.shape, val_data.shape
```

```python
y_train = train_data['expenses']
X_train = train_data[['age', 'bmi', 'smoker_yes',
       'region_northeast', 'region_northwest', 'region_southeast',
       'region_southwest', 'Num_children_2',
       'Num_children_3', 'Num_children_4']]

X_train = sm.add_constant(X_train)
model = sm.regression.linear_model.OLS(y_train, X_train).fit()

print(model.summary())
```

```python
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
```

```python
cols = ['age', 'bmi', 'smoker_yes',
       'region_northeast', 'region_northwest', 'region_southeast',
       'region_southwest', 'Num_children_2',
       'Num_children_3', 'Num_children_4']

np.sqrt(mean_squared_error(test_data['expenses'], model.predict( sm.add_constant(test_data[cols]))))
```
